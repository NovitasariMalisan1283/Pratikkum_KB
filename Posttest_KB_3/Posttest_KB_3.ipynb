```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('imdb.csv')
df.head()
```

## 1. GRAFIK BATANG

```python
# Membaca data dari file CSV
df = pd.read_csv('imdb.csv')

# Menghitung jumlah masing-masing peringkat IMDb
sc = df['Rate'].value_counts()

# Mengurutkan DataFrame berdasarkan nilai peringkat
sc = sc.sort_index()

# Mengatur ukuran gambar
plt.figure(figsize=(30, 30))

# Membuat grafik batang
bars = plt.bar(sc.index, sc, color=("red", "blue", "yellow"))

plt.title("Jumlah", fontsize=40, fontweight='bold')
plt.xlabel("Rate",fontsize=40, fontweight='bold')
plt.ylabel("Banyaknya", fontsize=40, fontweight='bold')

# Menambahkan teks besar dan tebal pada angka Rate dan banyaknya
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             str(int(bar.get_height())), ha='center', va='bottom', fontsize=16, fontweight='bold')

plt.show()
```

<p style="text-align: justify;"> Dari grafik batang di atas, kita dapat melihat bahwa dalam dataset IMDb yang digunakan dalam analisis ini, terdapat berbagai Rate IMDb yang tersebar untuk berbagai film yang ada. Grafik ini dengan jelas menunjukkan berapa banyak film yang ada dalam dataset ini, berdasarkan Rate IMDb masing-masing. Kita bisa melihat berbagai Rate IMDb dari yang terendah hingga yang tertinggi, dengan ukuran teks yang besar dan tebal pada sumbu x dan y, sehingga lebih mudah dibaca. Dengan visualisasi ini, kita dapat dengan cepat memahami sebaran Rate IMDb dalam dataset kita, yang tentunya sangat membantu ketika sedang menganalisis data.

## 2. PIE CHART

```python
# Membaca data dari file CSV
df = pd.read_csv('imdb.csv')

# Pie chart untuk atribut 'type' dengan efek explode
type_counts = df['Type'].value_counts()
explode = (0.1, 0)  # Menambahkan efek "explode" pada potongan pertama (0.1) dari pie chart
type_counts.plot(kind='pie', autopct='%1.1f%%', explode=explode)
plt.title('Pie Chart of Type Distribution')
plt.ylabel('')
plt.show()
```

<p style="text-align: justify;"> Dalam visualisasi diagram pie di atas, kita menganalisis distribusi tipe Movie dan tipe series dalam dataset IMDb. Dengan efek "explode" pada potongan pertama dari diagram pie, kita menyoroti tipe pertama yang ditemui dalam dataset. Hasilnya, dapat disimpulkan bahwa mayoritas film dalam dataset IMDb adalah tipe 'Movie', yang mendominasi dengan persentase sekitar 72.0%. Sementara itu, tipe series hanya menyumbang sekitar 28.0% dari seluruh dataset. Visualisasi ini memberikan gambaran singkat dan jelas tentang sebaran tipe film dan tipe series dalam dataset IMDb.

## 3. HEATMAP

```python
# Membaca data dari file CSV
df = pd.read_csv('imdb.csv')

# Mengonversi kolom 'Rate' ke tipe data float dan mengabaikan baris yang tidak dapat dikonversi
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
df.dropna(subset=['Rate'], inplace=True)

# Mengonversi kolom 'Votes' ke tipe data float dan mengabaikan baris yang tidak dapat dikonversi
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
df.dropna(subset=['Votes'], inplace=True)

# Mengonversi kolom 'Duration' ke tipe data float dan mengabaikan baris yang tidak dapat dikonversi
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
df.dropna(subset=['Duration'], inplace=True)

# Memilih atribut numerik untuk heatmap
numeric_columns = ['Rate', 'Votes', 'Duration']

# Menghitung matriks korelasi untuk atribut numerik
correlation_matrix = df[numeric_columns].corr()

# Membuat heatmap korelasi
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title('Heatmap Korelasi Antara Variabel Numerik', fontsize=16, fontweight='bold')
plt.xlabel('Variabel',fontsize=16, fontweight='bold')
plt.ylabel('Variabel',fontsize=16, fontweight='bold')
plt.show()
```

<p style="text-align: justify;"> Dalam visualisasi data tersebut, kita melakukan beberapa langkah untuk memproses data dari dataset IMDb. Pertama, kita membaca data dari file CSV dan mengubah beberapa kolom penting seperti peringkat (Rate), jumlah suara (Votes), dan durasi (Duration) menjadi bentuk numerik. 
Kemudian, kita melakukan analisis korelasi antara variabel-variabel numerik tersebut, yaitu Rate, Votes, dan Duration. Korelasi mengukur sejauh mana dua variabel berkaitan satu sama lai 

Hasil korelasi ini kemudian divisualisasikan dalam bentuk heatmap. Heatmap digunakan untuk memperlihatkan hubungan antar variabel dengan warna yang berbeda. Warna yang lebih terang menunjukkan hubungan yang lebih erat, sementara warna yang lebih gelap menunjukkan hubungan yang lemah atau bahkan tidak ada hubungan.

## 5. SCATTER PLOT


```python
# Membaca data dari file CSV
df = pd.read_csv('imdb.csv')

# Mengurutkan DataFrame berdasarkan kolom 'Rate' dari yang terbesar ke terkecil
df_sorted = df.sort_values(by='Rate', ascending=False)

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat scatter plot Date vs Rate
plt.figure(figsize=(10, 15))
sns.scatterplot(data=df_sorted, x='Date', y='Rate', alpha=0.5)
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Date vs Rate Scatter Plot (Sorted by Rate - Descending)', fontweight='bold')
plt.show()
```

<p style="text-align: justify;"> Visualisasi scatter plot "Date vs Rate" pada data IMDb memberikan gambaran mengenai hubungan antara tanggal perilisan dan peringkat film dalam dataset ini. Data diurutkan berdasarkan peringkat film secara menurun, sehingga kita dapat dengan jelas melihat film-film dengan peringkat tertinggi. Titik-titik pada plot mewakili setiap entri dalam dataset, dan transparansi titik-titik memungkinkan kita untuk mengidentifikasi kerumunan data yang lebih padat.

## 6. LINE PLOT

```python
# Membaca data dari file CSV
df = pd.read_csv('imdb.csv')

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Line plot Date vs Rate for the first 100 rows
plt.figure(figsize=(15, 17))
plt.plot(df['Date'][:100], df['Rate'][:100])  # Slicing the DataFrame for the first 100 rows
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Date vs Rate Line Plot (First 100 Rows)')
plt.xticks(rotation=60, horizontalalignment='right', fontsize=10)
plt.show()
```

<p style="text-align: justify;"> Dari grafik garis "Date vs Rate" yang dibuat untuk 100 baris pertama data, terlihat bahwa tidak ada pola yang jelas atau tren yang dapat dikenali pada hubungan antara tanggal (Date) dan peringkat (Rate) dalam dataset ini. Dalam 100 baris pertama data, peringkat (Rate) cenderung fluktuatif dan tidak menunjukkan pola naik atau turun yang konsisten seiring berjalannya waktu. Oleh karena itu, tidak dapat diambil kesimpulan yang kuat dari grafik ini terkait dengan hubungan antara tanggal dan peringkat dalam dataset.
