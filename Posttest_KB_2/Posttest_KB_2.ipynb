## 1. Tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
# Seleksi atribut yang bertipe angka
atribut_angka = df.select_dtypes(include=['number'])

# Hitung statistik yang diinginkan
statistik = atribut_angka.describe()
# Tambahkan perhitungan kuartil Q1, Q2 (median), dan Q3
statistik.loc['Q1'] = atribut_angka.quantile(0.25)
statistik.loc['Q2'] = atribut_angka.median()
statistik.loc['Q3'] = atribut_angka.quantile(0.75)

print(statistik)
```

## 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin

```python
import pandas as pd

# Membaca dataset
df = pd.read_csv('imdb.csv')

# Informasi jumlah record
jumlah_records = len(df)

# Informasi jumlah attribute
jumlah_attributes = len(df.columns)

# Informasi nama attribute
nama_attributes = df.columns.tolist()

# Informasi jumlah attribute dengan tipe angka
atribut_angka = df.select_dtypes(include=['number'])
jumlah_atribut_angka = len(atribut_angka.columns)

print("Nama Atribut:")
for i, nama in enumerate(nama_attributes, start=1):
    print(f"{i}. {nama}")
print("=" * 40)

# Contoh: Jumlah Atribut
print(f"Contoh:\nJumlah Atribut           = {jumlah_attributes}")
print("=" * 40)

# Tampilkan jumlah atribut dengan tipe angka dengan nomor urut
print("Atribut dengan Tipe Angka:")
for i, atribut in enumerate(atribut_angka.columns.tolist(), start=1):
    print(f"{i}. {atribut}")
print("=" * 40)

# Contoh: Jumlah Atribut dengan Tipe Angka
print(f"Contoh:\nJumlah Atribut dengan Tipe Angka = {jumlah_atribut_angka}")
print("=" * 40)

```

## 5. Tampilkan jumlah nilai dari semua attribute yang bernilai object.


```python
import pandas as pd

df = pd.read_csv('imdb.csv')

# Memilih kolom-kolom dengan tipe data objek (kategorikal)
objek = df.select_dtypes(include=['object'])

# Menghitung jumlah variasi (nilai unik) dalam setiap kolom kategorikal
jumlah_objek = objek.nunique()

# Membuat DataFrame untuk tabel variasi
tabel = pd.DataFrame({
    '= No. =': range(len(jumlah_objek)),
    '= Nama Atribut =': jumlah_objek.index, 
    '= Jumlah Nilai ': jumlah_objek.values})

print("===== Tabel  =====")
print(tabel.to_string(index=False, justify='center'))

```
